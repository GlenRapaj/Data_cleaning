```python
categorical_columns = [ 'aspiration', 'body-style', 'num-of-doors', 'drive-wheels', 'engine-location', 'engine-type', 'num-of-cylinders', 'fuel-system' ]
# df['fuel-system'].unique()
```

```python
for cc in categorical_columns:
    dumydf = pd.get_dummies( df[cc] )
    df = pd.concat( ( df, dumydf ), axis = 1 )
    df.drop( cc, axis = 1, inplace = True )
    del dumydf
```

```python
# normalize Data

df['wheel-base'] = ( df['wheel-base'] - df['wheel-base'].mean() ) /  df['wheel-base'].std()
df['width'] = ( df['width'] - df['width'].min() ) / ( df['width'].max()  - df['width'].min() )
df['height'] = ( df['height'] - df['height'].min() ) / ( df['height'].max()  - df['height'].min() )
df['bore'] = ( df['bore'] - df['bore'].min() ) / ( df['bore'].max()  - df['bore'].min() )
df['stroke'] = ( df['stroke'] - df['stroke'].min() ) / ( df['stroke'].max()  - df['stroke'].min() )
df['compression-ratio'] = ( df['compression-ratio'] - df['compression-ratio'].min() ) / ( df['compression-ratio'].max()  - df['compression-ratio'].min() )
df['city-mpg'] = ( df['city-mpg'] - df['city-mpg'].min() ) / ( df['city-mpg'].max()  - df['city-mpg'].min() )
df['highway-mpg'] = ( df['highway-mpg'] - df['highway-mpg'].min() ) / ( df['highway-mpg'].max()  - df['highway-mpg'].min() )
df['curb-weight'] = ( df['curb-weight'] - df['curb-weight'].mean() ) /  df['curb-weight'].std()
df['engine-size'] = ( df['engine-size'] - df['engine-size'].mean() ) /  df['engine-size'].std()
```
